minimumkan
$$f(x_1,x_2,x_3)=x_1^2+5x_2^2+25x_3^2-x_1-x_2-x_3$$
bentuk lain
$$f(x_1,x_2,x_3)=\begin{bmatrix} x_1&x_2&x_3 \end{bmatrix}\begin{bmatrix}
1 & 0 & 0 \\
0 & 5 & 0 \\
0 & 0 & 25 \end{bmatrix}\begin{bmatrix} x_1\\x_2\\x_3\end{bmatrix}-\begin{bmatrix} 1&1&1 \end{bmatrix}\begin{bmatrix} x_1\\x_2\\x_3 \end{bmatrix}$$
$$\nabla f(x)=\begin{bmatrix} 2x_1-1\\10x_2-1\\50x_3-1 \end{bmatrix}$$
$$\nabla^2f(x)=\begin{bmatrix} 2&0&0\\0&10&0\\0&0&50 \end{bmatrix}$$

```python
import numpy as np
```

```python
#Metode Steepest Descent

x=np.array([1,1,1]) #nilai awal
print("x0= ",x)
nablaf=np.array([-4*x[0]+x[1]-2,x[0]+2*x[1]+x[2]-6,x[1]+2*x[2]-5]) #menghitung nabla f(x) dengan x awal
Q=np.array([[-4,1,0],[1,2,1],[0,1,2]])
print("nabla f(x0)= ",nablaf)
print("norm(nabla f(x0))= ",np.linalg.norm(nablaf))
error=0.00001

iter=1
maxiter=100

while np.linalg.norm(nablaf)>error and iter<maxiter:
  alpha=(nablaf@nablaf)/(nablaf@Q@nablaf)
  x=x-alpha*nablaf
  print("\n","x",iter,"=",x)
  nablaf=np.array([-4*x[0]+x[1]-2,x[0]+2*x[1]+x[2]-6,x[1]+2*x[2]-5])
  print("nabla f(x",iter,")= ",nablaf)
  print("norm(nabla f(x",iter,")= ",np.linalg.norm(nablaf))
  iter=iter+1
```

```python
#Metode Newton

x=np.array([0,0,0]) #nilai awal
print("x0= ",x)
nablaf=np.array([-4*x[0]+x[1]-2,x[0]+2*x[1]+x[2]-6,x[1]+2*x[2]-5]) #menghitung nabla f(x) dengan x awal
nabla2f=np.array([[-4,1,0],[1,2,1],[0,1,2]]) #kebetulan konstan, tidak bergantung pada x
print("nabla f(x0)= ",nablaf)
print("norm(nabla f(x0))= ",np.linalg.norm(nablaf))
error=0.00001

iter=1
maxiter=100

while  np.linalg.norm(nablaf)>error and iter<maxiter:
  x=x-np.linalg.inv(nabla2f)@nablaf #memperbaharui x
  print("\n","x",iter,"= ",x)
  nablaf=np.array([-4*x[0]+x[1]-2,x[0]+2*x[1]+x[2]-6,x[1]+2*x[2]-5]) #menghitung nabla f(x) dengan x awal
  nabla2f=np.array([[-4,1,0],[1,2,1],[0,1,2]]) #kebetulan konstan, tidak bergantung pada x
  print("nabla f(x",iter,")= ",nablaf)
  print("norm(nabla f(x",iter,")= ",np.linalg.norm(nablaf))
  iter=iter+1

fx=np.array([2*x[0]**2+x[0]*x[1]+x[1]**2+x[1]*x[2]+x[2]**2-2*x[0]-6*x[1]-5*x[2]+2])
print(fx)
```

```python
#Metode Conjugate Gradient

x=np.array([0,0,0]) #nilai awal
print("x0= ",x)
nablaf=np.array([-4*x[0]+x[1]-2,x[0]+2*x[1]+x[2]-6,x[1]+2*x[2]-5]) #menghitung nabla f(x) dengan x awal
print("norm(nabla f(x0))= ",np.linalg.norm(nablaf))
nabla2f=np.array([[-4,1,0],[1,2,1],[0,1,2]])
d=-nablaf #menghitung d

iter=1
maxiter=100
error=0.000001

while  np.linalg.norm(nablaf)>error and iter<maxiter:
  alpha=-d@nablaf/(d@nabla2f@d)
  x=x+alpha*d
  print("x",iter,"= ", x)
  nablaf=np.array([-4*x[0]+x[1]-2,x[0]+2*x[1]+x[2]-6,x[1]+2*x[2]-5])
  print("norm(nabla f(x",iter,"))= ",np.linalg.norm(nablaf)) 
  d=-nablaf+( nabla2f@nablaf@d/(d@nabla2f@d)  )*d
  iter=iter+1
```
